```
def softmax(Z):
  e_Z = np.exp(Z)
  A = (e_Z)/(e_Z.sum(axis = 0))
  return A
# Z = np.array([1,2,2])
# print(softmax(Z).shape)

def getGrad(W,x,y):
  Z = W.T.dot(x)
  A = softmax(Z)
  # print(Z.shape)
  E = A-y
  # print(E.shape)
  return x.dot(E.T)

def Predict(x,w):
  print(w.shape)
  print(x.shape)
  Z = x.dot(w)
  probs = softmax(Z)
  print(probs)
  preds = np.argmax(probs,axis = 1)
  return preds

def softmax_regression(w_init,x,y,eta):
  W  = w_init
  iterations = 1000
  for i in range(0,iterations):
    # x = x.T
    # y = y.T
    grad = getGrad(W,x,y)
    # losses.append(loss)
    W = W - (eta * grad)
  return W

w_init = np.zeros([784,10])
# print(w_init)
learning_rate = 0.05
# print(x.shape,y.shape,w_init.shape)
w = softmax_regression(w_init,x,y,learning_rate)

label = Predict(x_test,w)
print(label)
# print(y_test.argmax(axis=1))
# y_test = y_test.argmax(axis=1)
acc = sum(label == y_test)/(10000.0)

print(acc)
```
